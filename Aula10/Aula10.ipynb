```python
strepisodios=df2["Episódios"][0]
lis=strepisodios.split("-")
lis2=[int(elem) for elem in lis]
lis2[1]-lis2[0]+1
```

```python
def geraduracao(string):
    lis=string.split("-")
    lis2=[int(elem) for elem in lis]
    return lis2[1]-lis2[0]+1
```

```python
teste=(df2["Episódios"].apply(geraduracao))
df2["Duração"]=teste
```
