```python
def substituir_meses(texto):
    meses_ingles = ["january", "february", "march", "april", "may", "june",
                    "july", "august", "september", "october", "november", "december"]
    meses_portugues = ["janeiro", "fevereiro", "março", "abril", "maio", "junho",
                       "julho", "agosto", "setembro", "outubro", "novembro", "dezembro"]

    texto = texto.lower()

    for mes_ing, mes_pt in zip(meses_ingles, meses_portugues):
        texto = texto.replace(mes_pt, mes_ing)

    return texto

# Testando a função
texto = "i was born on april 23, 1990 and my sister was born on december 5, 1995."
texto_traduzido = substituir_meses(texto)
print(texto_traduzido)
```

```python
from datetime import datetime

# Suponha que você tenha uma string de data no formato '1 de abril de 1997'
data_string = '1 de abril de 1997'
data_string_eng=substituir_meses(data_string)

# Use strptime para converter a string em um objeto datetime
data_objeto = datetime.strptime(data_string_eng, '%d de %B de %Y')

# Agora data_objeto é um objeto datetime representando a data 1 de abril de 1997
print(data_objeto)
```

```python
from datetime import datetime

# Obtendo a data e hora atual
agora = datetime.now()

# Formatando a data e hora no formato desejado
data_formatada = agora.strftime("%d de %B de %Y")

print(data_formatada)
```

```python
def gera_data(data_string):
  if data_string=="presente":
    data=datetime.now()
  else:
    data_string_eng=substituir_meses(data_string)
    data=datetime.strptime(data_string_eng, '%d de %B de %Y')

# Use strptime para converter a string em um objeto datetime
  return data

```

```python
for group in gb.groups:
  dftemp=gb.get_group(group)
  size=len(dftemp)
  delta=gera_data(dftemp["Final"].iloc[size-1])-gera_data(dftemp["Estreia"].iloc[0])
  print(("{Temporada} lasted for {days} days.").format(Temporada=group, days=delta.days))
```
